```python
#把資料叫進來
import pandas as pd
traindf = pd.read_csv("train.csv", encoding = "utf-8")
traindf
```

```python
testdf = pd.read_csv("test.csv", encoding = "utf-8")
testdf
```

```python
##測試1 用平均值

#判斷缺失值
traindf.isna().sum()
```

```python
#資料分布不均時 平均值不適用 所以要用中位數
med = traindf.median()
traindf = traindf.fillna(med)  ####填入中位數
traindf.isna().sum()
```

```python
##測試二 完全複製上面區段
testdf = testdf.fillna(med)  ####填入中位數
testdf.isna().sum()
```

```python
#缺失值缺類別就填最常出現的
most = traindf["Embarked"].value_counts().idxmax()
traindf["Embarked"] = traindf["Embarked"].fillna(most)
traindf.isna().sum()
```

```python
#電腦只接受數字型態的資料 所以要把文字轉成數字 而當不承認文字有大小關係 就要用one-hot encoding
dummy = pd.get_dummies(traindf["Embarked"]) ##準備出三個欄位的資料 準備做one hot encoding
traindf = pd.concat([traindf, dummy], axis=1) ##concat合併兩個表格
traindf
```

```python
dummy = pd.get_dummies(testdf["Embarked"])
testdf = pd.concat([testdf, dummy], axis=1)
testdf
```

```python
dummy = pd.get_dummies(traindf["Sex"]) ##準備出三個欄位的資料 準備做one hot encoding
traindf = pd.concat([traindf, dummy], axis=1) ##concat合併兩個表格
traindf
```

```python
dummy = pd.get_dummies(testdf["Sex"])
testdf = pd.concat([testdf, dummy], axis=1)
testdf
```

```python
def nameflow(s):
    mid = s.split(",")[-1].split(".")[0]
    mid = mid.replace(" ", "")
    return mid
name = traindf["Name"].apply(nameflow)
pd.crosstab(name, traindf['Survived']).T ##T 是矩陣反轉
```

```python
#pandas精隨 --> apply()功能
def nameflow(s):
    mid = s.split(",")[-1].split(".")[0]
    mid = mid.replace(" ", "")
    if mid == "Miss":
        return "Miss"
    elif mid == "Mr":
        return "Mr"
    elif mid == "Mrs":
        return "Mrs"
    else:
        return "Throw"

name = traindf["Name"].apply(nameflow)
dummy = pd.get_dummies(name) ##準備出三個欄位的資料 準備做one hot encoding
traindf = pd.concat([traindf, dummy], axis=1) ##concat合併兩個表格
traindf
```

```python
name = testdf["Name"].apply(nameflow)
dummy = pd.get_dummies(name) ##準備出三個欄位的資料 準備做one hot encoding
testdf = pd.concat([testdf, dummy], axis=1) ##concat合併兩個表格
testdf
```

```python
#看看有哪些欄位
traindf.columns
```

```python
#把不要的drop掉
traindf = traindf.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'Throw'], axis = 1)
testdf = testdf.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'Throw'], axis = 1)
traindf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#測試三 用決策樹
#自動做show() 只有jupyter可以用%
%matplotlib inline

#traindf.corr()是內建可用的相關係數圖 但很難看
plt.figure(figsize = (14, 14))
sns.heatmap(traindf.corr(), annot = True, cmap = "RdBu")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
trainx = traindf.drop(["Survived"], axis = 1)
trainy = traindf["Survived"]
# n_estimators代表你要幾棵樹
clf = RandomForestClassifier(max_depth=8, n_estimators = 28)
#多棵樹進行交叉驗證
np.average(cross_val_score(clf, trainx, trainy, cv=10))
```

```python
#透過貪婪搜索來找出最好的參數
from sklearn.model_selection import GridSearchCV
p = {
    "max_depth":range(5, 11),
    "n_estimators":range(20, 30)
}
clf = RandomForestClassifier()
s = GridSearchCV(clf, p, cv=5)
s.fit(trainx, trainy)
print(s.best_params_)
print(s.best_score_)
```
